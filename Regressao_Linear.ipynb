# **Conceitos**

A regressão linear, é útil para resolver problemas simples de análise de dados preditivas quando se uma variável-alvo numérica.

Uma regressão linear, basicamente, é uma técnica matemática de aproximação de algo por uma reta. É utilizado técnicas estatísticas para buscar qual seria o erro aproximado se fosse utilizar aquela reta para prever o futuro.

Como é calculada uma regresão linear:

A ideia básica da regressão linear passa pela premissa de que duas variáveis tenham uma relação linear entre si. Que pode ser visto através de um gráfico de dispersão entre duas variáveis, mas também gerar uma métrica que ajuda a entender a relação linear entre duas variáveis. Essa métrica é o coeficiente de correlação linear.

A regresão busca uma reta que melhor se ajusta em um conjunto de pontos dados. Essa reta pode ser calculada através da equação da reta que é:

y = Ⲁx + ϐ + ɛ

Onde: y e x: valores dos pontos no sentido vertical e horizontal, respectivamente.

Ⲁ: coeficiente angular da reta, ou seja, este valor diz se uma reta está "apontando" para cima, para baixo e qual é a sua inclinação.

ϐ: coeficiente linear ou de intercepto, que nada mais é do que o valor que a reta assume quando intercepta o eixo y (onde x=0).

ɛ: erros de aproximação da reta associados a cada ponto.

Se encontrarmos valores com um alto coeficiente de correlação, poderemos supor que um influencia em outro (relação causa-efeito). Assim sendo, a partir de estimativas futuras da variável independente, calcularemos a previsão da variável dependente.

**Regressão Linear em Python**

```python
# 1 - Carga das bibliotecas
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
# Função para calcular o erro de previsão
from sklearn.metrics import mean_squared_error

# 2 - Carga dos dados
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP/master/dados/wine/red_wine4_30.csv", encoding='latin-1')

# 3 - Preparação dos dados

# 4 - Separação das amostras de treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(df[['wine_year', 'wine_price']], df['wine_rating'],test_size=0.3266, random_state=33)

# 5 - Construção do modelo
mod = LinearRegression()
mod.fit(x_treino,y_treino)

# Cálculo das estatísticas da regressão linear
import statsmodels.api as sm
mod = sm.OLS(y_treino, x_treino).fit()
print(mod.summary())

# 6 - Avaliação do modelo
y_prev = mod.predict(x_teste)
mean_squared_error(y_teste, y_prev)
```

```python
# 1 - Carga das bibliotecas
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
# 2 - Carga dos dados
df = pd.read_csv("https://raw.githubusercontent.com/diogenesjusto/FIAP-On-Line-Graduation/main/datasets/pib.csv")

# 3 - Preparação dos dados

# 4 - Separação das amostras de treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(df[['BRP']], df['PIB'],test_size=0.3266, random_state=33)

# 5 - Construção do modelo
mod = LinearRegression()
mod.fit(x_treino,y_treino)

# Cálculo das estatísticas da regressão linear
import statsmodels.api as sm
mod = sm.OLS(y_treino, x_treino).fit()
print(mod.summary())

# 6 - Avaliação do modelo
y_prev = mod.predict(x_teste)
mean_squared_error(y_teste, y_prev)
```

R-squared: o R quadrado é uma métrica que sinaliza o quanto uma modelo linear está conseguindo representar os dados. O melhor valor de R2 seria mais próximo de 1 possível (100%). Neste caso 0,99 é um valor bem alto.

Std err: o erro padrão é uma métrica que sinaliza qual é o erro médio (distância) do modelo para os pontos. Um erro baixo, pequeno, será mais adequado que um erro grande.

P>|t|: o p-valor é a métrica que sinaliza quando menor, mais a significância estatística. A significância estatística será 1 (ou 100%) menos o p-valor. Um p-valor na ordem de 0,05 (5%) significa dizer que a significância estatística do meu modelo será da ordem de 0,95 (95%). Podemos interpretar da seguinte forma: se realizarmos 100 previsões, 95 estão dentro de uma faixa de erro médio, sendo que 5 poderão ser muito maiores ou muito menores.

coef: Este é literalmente o coeficiente da regressão linear. Uma regressão com somente uma variável preditora, portanto temos somente um coeficiente.
